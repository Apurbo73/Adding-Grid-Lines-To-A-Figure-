```python
import matplotlib.pyplot as plt
import pandas as pd
#Data:
data ={
    "Cricket_bat": ["SG", "CA","SS"],
    "MRP": [ 1000,2000,3000],
    "WEIGHT": [1100,1300,1250]

}
dataFrame= pd.DataFrame(data)
# PLot a line graph
plt.plot(dataFrame["Cricket_bat"],dataFrame["MRP"], dataFrame["WEIGHT"])
# Add a grid Line:
plt.grid()
plt.show()
```
